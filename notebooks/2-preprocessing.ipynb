```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
train = pd.read_csv('../data/train.csv')
test = pd.read_csv('../data/test.csv')
```

```python
vectorizer = TfidfVectorizer()
```

```python
X_train = vectorizer.fit_transform(train['comment']).toarray()
```

```python
X_test = vectorizer.transform(test['comment']).toarray()
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
```

```python
models, predictions = clf.fit(X_train, X_test, train['isHate'], test['isHate'])
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
train['comment'] = train['comment'].apply(
    lambda x: ' '.join([
        tok.lemma_.lower() for tok in nlp(x)
        if not tok.is_punct
    ])
)
```

```python
train
```

```python
X_train = vectorizer.fit_transform(train['comment']).toarray()
X_test = vectorizer.transform(test['comment']).toarray()
```

```python
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
```

```python
models, predictions = clf.fit(X_train, X_test, train['isHate'], test['isHate'])
```
